```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
nivel_estudios = []
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)

nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
```

```python
# DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

```python
import pandas as pd

## datos 
datos = (['Secundaria'] *110 + ['Bachillerato'] *172 + ['Grado'] *145 + ['Master'] *52 + ['Doctorado'] *21)

##crear data frame
df = pd.DataFrame(datos, columns=['Nivel_Estudios'])

# para utilizar datos ramdom 
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
# Ver las primeras filas
df.head()
##print(datos)
```

```python
# Generar tabla de frecuencias
tabla_frecuencias = df['Nivel_Estudios'].value_counts().to_frame('Frecuencia_Absoluta')

#  frecuencia relativa 
tabla_frecuencias['Porcentaje'] = (df['Nivel_Estudios'].value_counts(normalize=True) * 100).round(2)

print("Resumen Estadístico de Variables Cualitativas:")
display(tabla_frecuencias)
```

```python
plt.style.use('ggplot')
# Graficar
ax = tabla_frecuencias['Frecuencia_Absoluta'].plot(kind='bar', 
                                                   figsize=(10, 5), 
                                                   color='skyblue', 
                                                   edgecolor='black')

plt.title('Distribución de Niveles de Estudio')
plt.xlabel('Grado Académico')
plt.ylabel('Cantidad de Personas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pastel
tabla_frecuencias['Frecuencia_Absoluta'].plot(kind='pie', 
                                              figsize=(8, 8), 
                                              autopct='%1.2f%%', 
                                              startangle=90,
                                              cmap='Pastel1')

plt.title('Proporción por Nivel Educativo')

plt.show()
```
